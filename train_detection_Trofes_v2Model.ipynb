```python
# 1. Mount Drive & Install Library
from google.colab import drive
from ultralytics import YOLO
import os

drive.mount('/content/drive',force_remount=True)


# 2. Persiapan Folder & Ekstrak Dataset (Nama File yang Benar)
# Menggunakan nama file dari gambar yang kamu kirim tadi
dataset_zip = "/content/drive/MyDrive/Trofes_Detection_dataset3.v4-tes-4.yolov8.zip"
target_folder = "/content/dataset"

if not os.path.exists(target_folder):
    os.makedirs(target_folder)
    # Pastikan nama file ZIP di bawah ini sesuai dengan yang ada di Drive-mu
    !unzip -q "{dataset_zip}" -d {target_folder}

    # Cek apakah unzip berhasil
    if os.path.exists(f"{target_folder}/data.yaml"):
        print("BERHASIL! Dataset dan data.yaml sudah siap di akun kedua.")
    else:
        print("GAGAL! File ZIP tidak ditemukan atau nama file salah.")
        print("Daftar file di Drive-mu:", os.listdir('/content/drive/MyDrive/'))
else:
    print("‚Ñπ Folder dataset sudah ada, tidak perlu unzip ulang.")

# 3. Resume Training dari Titik Terakhir (Epoch 9)
# Pastikan path folder project ini sudah benar di Drive akun kedua
path_last = '/content/drive/MyDrive/Trofes_Project/yolov8m_Detection_colab/weights/last.pt'

if os.path.exists(path_last):
    print("Menghubungkan ke 'last.pt',Melanjutkan training Trofes!")
    model = YOLO(path_last)
    # resume=True akan otomatis mencari data.yaml yang baru saja di-unzip
    model.train(resume=True)
else:
    print("ERROR: File 'last.pt' tidak ditemukan!")
    print("Pastikan folder 'Trofes_Project' sudah di-share sebagai Editor dan di-Add Shortcut.")
```
