```python
from ultralytics import YOLO
import os

# 1. Load model Medium (m) dari lokasi yang kamu sebutkan
model_path = "/content/drive/MyDrive/Trofes_projectV2/yolov8m_train/weights/best.pt"
model_m = YOLO(model_path)

print(" Memulai Evaluasi Model Medium (Trofes V2)...")

# 2. Jalankan evaluasi pada data 'test' (biar jujur, bukan data training)
# split='test' memastikan kita menguji model pada data yang benar-benar asing
metrics_m = model_m.val(data="/content/dataset/data.yaml", split='test')

print("\n" + "="*30)
print("       HASIL EVALUASI MODEL M")
print("="*30)
print(f"Akurasi (mAP50)     : {metrics_m.box.map50:.4f}")
print(f"Akurasi Ketat (mAP50-95) : {metrics_m.box.map:.4f}")
print(f"Kecepatan Inferences: {metrics_m.speed['inference']:.2f} ms per image")
print("="*30)

# 3. Simpan hasil evaluasi visual (Confusion Matrix, dll) ke folder khusus biar ga hilang
print("Evaluasi selesai. Cek folder 'runs/detect/val' untuk melihat Confusion Matrix terbaru.")
```

```python
import shutil
import os

# 1. Tentukan folder sumber dan tujuan di dalam Trofes_projectV2
source_dir = "/content/runs/detect/val5"
dest_dir = "/content/drive/MyDrive/Trofes_projectV2/Dokumentasi_Medium"

# 2. Buat folder tujuan di Drive
if not os.path.exists(dest_dir):
    os.makedirs(dest_dir)
    print(f"Folder {dest_dir} berhasil dibuat.")

# 3. Pindahkan semua file hasil evaluasi
if os.path.exists(source_dir):
    files = os.listdir(source_dir)
    for f in files:
        shutil.copy(os.path.join(source_dir, f), os.path.join(dest_dir, f))
    print(f" Aman! {len(files)} file evaluasi sudah disalin ke Drive.")
else:
    print("Folder hasil evaluasi tidak ditemukan. Pastikan evaluasi tadi sudah selesai running.")
```
